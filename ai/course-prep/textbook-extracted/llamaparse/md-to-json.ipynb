```python
def read_md_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            md_content = file.read()
        return md_content
    except:
        return -1

file_path = "./kang_math_5th_1st_extracted.md"  
markdown_content = read_md_file(file_path)
```

```python
import json
import re

def parse_markdown(markdown_text):
    """Parse the Markdown text and convert it to a JSON object."""
    lines = markdown_text.split('\n')
    json_obj = []
    current_block = None

    for line in lines:
        if line.startswith('#'):
            header_level = len(re.match(r'#+', line).group(0))
            header_text = line[header_level:].strip()
            json_obj.append({"type": "header", "level": header_level, "text": header_text})
        elif line.startswith('- ') or line.startswith('* '):
            item_text = line[2:].strip()
            if current_block and current_block['type'] == 'list':
                current_block['items'].append(item_text)
            else:
                current_block = {"type": "list", "items": [item_text]}
                json_obj.append(current_block)
        elif line.strip() == '':
            current_block = None
        else:
            if current_block and current_block['type'] == 'paragraph':
                current_block['text'] += ' ' + line.strip()
            else:
                current_block = {"type": "paragraph", "text": line.strip()}
                json_obj.append(current_block)
    
    return json_obj

def markdown_to_json(markdown_text):
    """Convert Markdown text to a JSON string."""
    json_obj = parse_markdown(markdown_text)
    return json.dumps(json_obj, indent=2)
```

```python
json_output = markdown_to_json(markdown_content)
print(json_output)
```

```python
def unicode_to_chinese(unicode_str):
    return unicode_str.encode('ascii').decode('unicode_escape')

print(unicode_to_chinese(json_output))
```
